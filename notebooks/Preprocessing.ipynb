```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, Normalizer

# Sample dataset creation
data = {
    'Feature1': [10, 20, 30, 40, 50],
    'Feature2': [100, 200, 300, 400, 500],
    'Feature3': [1000, 2000, 3000, 4000, 5000]
}

df = pd.DataFrame(data)

# Display the original data
print("Original Data:")
print(df)
```

```python
# Standardization (Z-score normalization)
scaler = StandardScaler()
standardized_data = scaler.fit_transform(df)
```

```python
standardized_data
```

```python

df_standardized = pd.DataFrame(standardized_data, columns=df.columns)

print("\nStandardized Data:")
print(df_standardized)
```

image.png

```python
# Min-Max Normalization (Scaling to [0, 1])
min_max_scaler = MinMaxScaler()
min_max_scaled_data = min_max_scaler.fit_transform(df)
df_min_max_scaled = pd.DataFrame(min_max_scaled_data, columns=df.columns)

print("\nMin-Max Scaled Data:")
print(df_min_max_scaled)
```

image.png

```python
# Robust Scaling (Scaling using median and IQR)  -1  to +1
robust_scaler = RobustScaler()
robust_scaled_data = robust_scaler.fit_transform(df)
df_robust_scaled = pd.DataFrame(robust_scaled_data, columns=df.columns)

print("\nRobust Scaled Data:")
print(df_robust_scaled)
```

image.png

```python
# Normalization (L2 Norm)    0 to 1
normalizer = Normalizer()
normalized_data = normalizer.fit_transform(df)
df_normalized = pd.DataFrame(normalized_data, columns=df.columns)

print("\nNormalized Data:")
print(df_normalized)
```
